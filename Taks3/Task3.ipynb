```python
from matplotlib import pyplot
import pandas as p

dataset=p.read_csv('diabetes.csv')
row=dataset.shape[0]
print(f"Number of rows  {row}")
print()
print("-------------------------Information-----------------------")
print()
print(dataset.info())
print()

print("----------------------------Summary----------------------------")
print()
print(dataset.describe())
print()
# histogram

dataset.hist()


pyplot.tight_layout()      # to restrain from overlapping

pyplot.show()
print()

dataset.plot() # density
pyplot.tight_layout()
pyplot.show()

print()
#box plot
dataset.plot(kind='box',subplots=True,layout=(3,3),sharex=False,sharey=False)
pyplot.tight_layout()
pyplot.show()
```
